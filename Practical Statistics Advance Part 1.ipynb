```python
#2.  Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range of values (e.g., like a fair 6-sided die)
low = 1
high = 7

# Create a discrete uniform distribution
dist = randint(low, high)

# Values the variable can take
x = np.arange(low, high)

# Probability Mass Function values
pmf_values = dist.pmf(x)

# Plot PMF using stem
plt.stem(x, pmf_values, basefmt=" ")  # Removed use_line_collection
plt.title("PMF of Discrete Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        return 0  # Bernoulli is only defined for 0 or 1
    return p if x == 1 else 1 - p

# Example
p = 0.7
print("P(X=1):", bernoulli_pdf(1, p))
print("P(X=0):", bernoulli_pdf(0, p))
```

```python
#6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
#  uniform distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters for discrete uniform distribution
low = 1
high = 7  # like a die roll: 1 through 6

# Create distribution
dist = randint(low, high)

# X values and CDF values
x = np.arange(low - 1, high + 1)
cdf = dist.cdf(x)

# Plot CDF
plt.step(x, cdf, where='post')
plt.title("CDF of Discrete Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.grid(True)
plt.show()
```

```python
#9.  Write a Python function to calculate Z-scores from a dataset and plot them.

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

# Example data
data = np.random.normal(50, 10, 1000)
z_scores = calculate_z_scores(data)

# Plot Z-scores
plt.hist(z_scores, bins=30, edgecolor='black', density=True)
plt.title("Z-scores Distribution")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
#10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

# Generate non-normal data (exponential)
population = np.random.exponential(scale=2, size=10000)

# Parameters
sample_size = 30
num_samples = 1000
sample_means = []

# Take multiple samples and store their means
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size)
    sample_means.append(np.mean(sample))

# Plot histogram of sample means
plt.hist(sample_means, bins=30, edgecolor='black', density=True)
plt.title("Central Limit Theorem (Means of Exponential Distribution)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
#12 . Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).
from scipy.stats import norm

def plot_standard_normal():
    x = np.linspace(-4, 4, 1000)  # range of Z-scores
    pdf = norm.pdf(x, loc=0, scale=1)  # standard normal: mean=0, std=1

    plt.plot(x, pdf, color='blue', lw=2)
    plt.title("Standard Normal Distribution (Z ~ N(0, 1))")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.grid(True)
    plt.show()

# Call the function
plot_standard_normal()

```

```python
#13. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
#     distribution.
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    return (x - mean) / std

# Example data
x = 72
mean = 70
std = 5

# Calculate Z-score
z = calculate_z_score(x, mean, std)
print("Z-score:", z)

# Plot standard normal distribution and mark Z-score
x_vals = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x_vals)

plt.plot(x_vals, pdf, label="Standard Normal Curve")
plt.axvline(z, color='red', linestyle='--', label=f"Z = {z:.2f}")
plt.fill_between(x_vals, 0, pdf, where=(x_vals < z), color='skyblue', alpha=0.4)
plt.title("Z-score Comparison to Standard Normal Distribution")
plt.xlabel("Z")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#15.  Implement hypothesis testing using Z-statistics for a sample dataset.
import numpy as np
from scipy.stats import norm

# Population parameters
mu = 50        # population mean
sigma = 10     # population std dev

# Sample data
sample = np.array([52, 55, 48, 51, 53, 49, 47, 56, 54, 50])
n = len(sample)
sample_mean = np.mean(sample)

# Z-statistic
z = (sample_mean - mu) / (sigma / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))  # two-tailed test

print("Sample Mean:", sample_mean)
print("Z-statistic:", z)
print("P-value:", p_value)

# Conclusion
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
#16.  Create a confidence interval for a dataset using Python and interpret the result.

# Sample data
data = np.array([102, 98, 100, 101, 99, 97, 103, 104, 96, 100])
sample_mean = np.mean(data)
n = len(data)
std_dev = np.std(data, ddof=1)  # sample std dev

# Z critical value for 95% CI
z_critical = norm.ppf(0.975)  # two-tailed

# Margin of error
margin = z_critical * (std_dev / np.sqrt(n))

# Confidence interval
ci_lower = sample_mean - margin
ci_upper = sample_mean + margin

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#17.  Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
# Generate normal data
np.random.seed(42)
data = np.random.normal(loc=70, scale=5, size=50)  # μ=70, σ=5

# Sample mean and std dev
sample_mean = np.mean(data)
std_dev = np.std(data, ddof=1)
n = len(data)

# Confidence interval (95%)
z_critical = norm.ppf(0.975)
margin = z_critical * (std_dev / np.sqrt(n))
ci_lower = sample_mean - margin
ci_upper = sample_mean + margin

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

```

```python
#19 Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.
from scipy.stats import poisson

# Poisson parameters
mu = 3  # mean number of events
x = np.arange(0, 11)  # 0 to 10 events

# CDF values
cdf = poisson.cdf(x, mu)

# Display values
for val, prob in zip(x, cdf):
    print(f"P(X ≤ {val}) = {prob:.4f}")

# Plot
plt.step(x, cdf, where='post', color='green')
plt.title("CDF of Poisson Distribution (μ = 3)")
plt.xlabel("x (number of events)")
plt.ylabel("Cumulative Probability")
plt.grid(True)
plt.show()
```

```python
#21.  Write a Python program to compare the standard deviations of two datasets and visualize the difference.
import numpy as np
import matplotlib.pyplot as plt

# Generate datasets
data1 = np.random.normal(50, 5, 1000)   # smaller spread
data2 = np.random.normal(50, 15, 1000)  # larger spread

# Calculate standard deviations
std1 = np.std(data1)
std2 = np.std(data2)

print(f"Standard Deviation of data1: {std1:.2f}")
print(f"Standard Deviation of data2: {std2:.2f}")

# Plot
plt.hist(data1, bins=30, alpha=0.6, label='Data1 (std=5)')
plt.hist(data2, bins=30, alpha=0.6, label='Data2 (std=15)')
plt.title("Comparison of Standard Deviations")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#22.  Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.
import scipy.stats as stats

# Generate normal data
data = np.random.normal(100, 15, 1000)

# Range
range_val = np.max(data) - np.min(data)

# Interquartile Range (IQR)
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

print(f"Range: {range_val:.2f}")
print(f"IQR: {iqr:.2f}")
```

```python
#23.  Implement Z-score normalization on a dataset and visualize its transformation.

# Original data
data = np.random.normal(60, 10, 1000)

# Z-score normalization
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

# Plot
plt.hist(data, bins=30, alpha=0.6, label='Original Data')
plt.hist(z_scores, bins=30, alpha=0.6, label='Z-score Normalized')
plt.title("Z-score Normalization Comparison")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#23  Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
#   distribution.

from scipy.stats import skew, kurtosis

# Generate normal data
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculate
skewness = skew(data)
kurt = kurtosis(data)  # default is Fisher's (normal = 0)

print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis: {kurt:.4f}")
```
